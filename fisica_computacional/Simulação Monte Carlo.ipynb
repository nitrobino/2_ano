```python
import numpy as np
import pylab as p
from IPython.display import Audio, display
from time import time
from scipy import optimize 
```

```python
def monte_carlo_1(M,x0,t):
    N = int(M/len(x0))
    arr = np.zeros((N,len(x0)))
    
    for j in range(t):
        x = np.random.normal(x0,1,(len(x0)))
        p = np.random.rand(len(x0))
        v = np.where(p < rho(x)/rho(x0))
        x0[v] = np.copy(x[v],float)
    for i in range(N):
        x = np.random.normal(x0,1,(len(x0)))
        p = np.random.rand(len(x0))
        v = np.where(p < rho(x)/rho(x0))
        x0[v] = np.copy(x[v])
        arr[i] = np.copy(x0)
        
    return np.reshape(arr,(N*len(x0)))
```

```python
def rho(x):                             # cauchy
    return 1/np.pi*1/(1+(x-1)**2)
def cauchy(x):
    return 1/np.pi*1/(1+(x-1)**2)
t1 = time()
x0 = np.array([2,4],float)
x = monte_carlo_1(100000,x0,100)
t2 = time()
print(t2 - t1)

x2 = np.linspace(-20,20,1000)
p.hist(x, bins = 'auto', range = (-10,10), density = 'true')
p.plot(x2,cauchy(x2))
p.xlim(-10,10)
p.show()
```

```python
def rho(x):             # gaussianas
    u1 = 6
    u2 = 14
    return (np.exp(-(x-u1)**2) + np.exp(-(x-u2)**2))*np.sqrt(1/np.pi)/2

t1 = time()
x0 = np.array([2,15],float)
x = monte_carlo_1(1000000,x0,100)
t2 = time()
print(t2 - t1)

x2 = np.linspace(0,20,1000)
p.hist(x, bins = 'auto', range = (0,20), density = 'true')
p.plot(x2,rho(x2))
p.xlim(0,20)
p.show()
```
