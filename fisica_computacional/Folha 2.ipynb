```python
import pylab as p
import numpy as np
import math 


data = np.loadtxt("sunspots.txt",float)
x = data[:,0]
y = data[:,1]
yav = []
for i in range(len(y) ):
    yav.append((1/11)*sum(y[i-5:i+6]))
            
p.xlabel('anos')
p.ylabel('sunspots')
p.xlim(0,1000)
p.plot(x,y)
p.plot(x,yav)
p.show()
```

```python
M = 24*np.pi
theta = np.linspace(0,M,10000)
r = np.exp(np.cos(theta))-2*np.sin(4*theta) + (np.sin(theta/12))**5
x = r*np.cos(theta)
y = r*np.sin(theta)

plot(x,y)
show()
```

```python

matriz1 = np.zeros((100,100))
matriz2 = np.zeros((100,100))
x1=40
y1=50
x2=60
y2=50
comp=10
A=5
k = 2*math.pi/comp
for i in range(100):
    for j in range(100):
        matriz2[i][j] = A*np.sin(k*np.sqrt((j-x2)**2+(i-y2)**2))
        matriz1[i][j] = A*np.sin(k*np.sqrt((j-x1)**2+(i-y1)**2))
p.imshow(matriz2 + matriz1)

p.show()
```

```python
import pylab as p
import numpy as np
import time as t
t1 = t.time()
N = 1000
M = 100
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)

matriz = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        z = np.zeros(2)
        a = 0
        for k in range(1,M):
            z = np.array([z[0]**2-z[1]**2,2*z[0]*z[1]]) + np.array([x[j],y[i]])
            if sum(z*z) > 4:
                break
            matriz[i][j] = np.log(k)
print(t.time()-t1)
p.imshow(matriz, origin = 'lower',extent=[-2,2,-2,2] )
p.jet()
p.colorbar()
p.show()
            
            
```
