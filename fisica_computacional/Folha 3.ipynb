```python
def fatorial(n):
    if n == 0:
        return 1
    else:
        return n*fatorial(n-1)
fatorial(200)
```

```python
def fatorialio(n):
    if n == 0:
        return 1.0
    else:
        return float(n*fatorialio(n-1))
fatorialio(200)
```

```python
import numpy as np 
def poli_1(a,b,c):
    return [(-b+np.sqrt(b**2-4*a*c))/(2*a),(-b-np.sqrt(b**2-4*a*c))/(2*a)]

def poli_2(a,b,c):
    return [2*c/(-b-np.sqrt(b**2-4*a*c)),2*c/(-b+np.sqrt(b**2-4*a*c))]
print(poli_1(0.001,1000,0.001))
print(poli_2(0.001,1000,0.001))
    
```

```python
def poli_3(a,b,c):
    if b>=0:
        return [poli_2(a,b,c)[0],poli_1(a,b,c)[1]]
    else:
        return [poli_1(a,b,c)[0],poli_2(a,b,c)[1]]
print (poli_3(0.001,1000,0.001))
```

```python
def f(x):
    return x*(x-1)
def derivada(x,d):
    return (f(x+d)-f(x))/d
derivada(1,10**(-6))
```

```python
from pylab import plot,show
x = np.linspace(10**(-6),10**(-14),100)
y = derivada(1,x)
print(y)
plot(x,y)
show()
```

```python
from pylab import xlim,ylim,plot,show
x =np.linspace(10**(-14),10**(-6),100, False)
y = []
for i in x:
    y.append(abs(1-derivada(1,i)))
    
plot(x,y,'.')
ylim(-0.,0.000001)
show()
```
